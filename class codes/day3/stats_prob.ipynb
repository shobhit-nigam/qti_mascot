```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# y = 2x+1
# linear equation

x = np.linspace(-10, 10, 100)
y = 2 * x + 1

# plotting the linear equation
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="y = 2x+1")
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# actual -> 	3, -0.5, 2, 7
# predicted	2.5, 0, 2, 8

#actual values
y_true = np.array([3, -0.5, 2, 7])


#predicted values 
y_pred = np.array([2.5, 0, 2, 8])

# mse
mse = np.mean((y_true-y_pred) ** 2)

print(f'MSE = {mse}')
```

```python
# logistic function
def logistic(xa):
    ya = 1 / ( 1 + np.exp(-x))
    return ya

x = np.linspace(-10, 10, 100)
y = logistic(x)


# plotting the logistic function
plt.figure(figsize=(8, 6))
plt.plot(x, y, label="logistic function")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
